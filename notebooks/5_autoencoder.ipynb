```python
import joblib
from joblib import load

df = joblib.load('dataset joblib files\dataset_autoencoder.joblib')
```

```python
df1 = df[df['is_laundering']==0]

df1.head()
```

```python
df1.is_laundering.value_counts()
```

```python
df2 = df1.copy()
columns_to_change = ['from_bank', 'account', 'to_bank', 'account_1']

for col in columns_to_change:
    freq = df2[col].value_counts(normalize=True)
    df2[col] = df2[col].map(freq)
```

```python
df2.head()
```

```python
df2.info()
```

Robust scaling handles outliers better than normalization or standardization by minimizing their influence while scaling numerical features. This ensures stable transformations and preserves the integrity of the data distribution for analysis and modeling.

```python
robust_scaler = RobustScaler()
columns_to_scale = ['amount_received', 'amount_paid']

df3 = df2.copy()

df3[columns_to_scale] = robust_scaler.fit_transform(df3[columns_to_scale])
df3[columns_to_scale] = robust_scaler.transform(df3[columns_to_scale])

df3.head()
```

```python
# for smote applied one
import joblib
from joblib import dump

dump((df3), 'df_auto_fixed.joblib')
```
